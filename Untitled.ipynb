# New data

```python
import numpy as np

fully_stocked= 100
n_hospitals = 60

mat = np.zeros((n_hospitals, n_hospitals))

for i in range(n_hospitals):
    fully_stocked = 100
    for j in range(n_hospitals):
        rm = np.random.randint(1, 25, 1)
        process = np.random.binomial(1, 1/fully_stocked, 1)
        
        
        if j == 0:
            mat[i, j] = fully_stocked   
        elif fully_stocked <= 10 or fully_stocked <= rm:
            
            if process != 0:
                fully_stocked = 100
            else:
                fully_stocked *= 2
                
            mat[i, j] = fully_stocked
        else:
            fully_stocked -= rm
            mat[i,j] = fully_stocked
            
```
